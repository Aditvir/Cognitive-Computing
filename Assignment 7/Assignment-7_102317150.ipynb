```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rollNumber = 102317150

np.random.seed(rollNumber)

salesData = np.random.randint(1000, 5000, size=(12, 4))

columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
salesDF = pd.DataFrame(salesData, columns=columns, index=months)

salesDF.head()
salesDF.describe()
```

```python
growthRates = salesDF.pct_change() * 100 #calculates the percentage change between the current and previous element

salesDF['Total Sales'] = salesDF.sum(axis=1)

salesDF['Growth Rate'] = salesDF['Total Sales'].pct_change() * 100

if rollNumber % 2 == 0:
    salesDF['Electronics'] = salesDF['Electronics'] * 0.9
else:
    salesDF['Clothing'] = salesDF['Clothing'] * 0.85
```

```python
plt.figure(figsize=(10, 6))

for category in columns:
    plt.plot(salesDF.index, salesDF[category], label=category)

plt.title('Monthly Sales Trends for Each Category')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=salesDF[columns])
plt.title('Sales Distribution for Each Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floors = np.floor(a)
ceilings = np.ceil(a)
truncated = np.trunc(a)
rounded = np.round(a)

print("Floor values:", floors)
print("Ceiling values:", ceilings)
print("Truncated values:", truncated)
print("Rounded values:", rounded)
```

```python
def swap_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

my_list = [10, 20, 30, 40, 50]
index1 = 1
index2 = 3

swapped_list = swap_elements(my_list, index1, index2)
print(swapped_list)
```

```python
def swap_elements_in_set(s, index1, index2):
    s_list = list(s)
    s_list[index1], s_list[index2] = s_list[index2], s_list[index1]
    return set(s_list)

my_set = {10, 20, 30, 40, 50}
index1 = 1
index2 = 3

swapped_set = swap_elements_in_set(my_set, index1, index2)
print(swapped_set)
```
